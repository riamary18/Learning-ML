## Dot Product (Example)

```python
np.random.seed(0)
sales_amount=np.random.randint(0,20,size=(5,3))
sales_amount
```

```python
#Create weekly sales dataframe
```

```python
import pandas as pd
weekly_sales=pd.DataFrame(sales_amount,index=["Mon","Tue","Wed","Thurs","Fri"],columns=["Almond butter","Peanut butter","Cashew butter"])
weekly_sales
```

```python
#Create prices array
prices=np.array([10,8,12])
prices
```

```python
butter_prices=pd.DataFrame(prices.reshape(1,3),index=["Prices"],columns=["Almond butter","Peanut butter","Cashew butter"])
butter_prices
```

```python
prices.shape
```

```python
total=prices.dot(sales_amount.T)
total
```

```python
total1=np.dot(weekly_sales,butter_prices.T)
total1.T
```

```python
weekly_sales["Total"]=total
weekly_sales
```
