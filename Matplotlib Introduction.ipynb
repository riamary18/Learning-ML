```python
# Make a bar plot from dictionary
prices={"almond":10,"peanut":12,"cashew":14}
fig,ax=plt.subplots()
ax.bar(prices.keys(),prices.values())
ax.set(title='Nut sales',
      xlabel='nuts',
      ylabel='price');
```

### Which one should you use? (pyplot vs 00 method)

```python
#When plotting something quickly, you can use pyplot method
#When plotting something more advanced, use the 00 method
```

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
#Pyplot method
over_50.plot(kind="scatter",x="age",y="chol",c="target");
```

```python
#00 method mixed with pyplot method
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
#ax.set_xlim([45,100])
```

```python
#00 method from scratch
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
ax.set(title="Heart Disease",xlabel='age',ylabel='chol')
ax.legend(*scatter.legend_elements(),title='Target')
ax.axhline(over_50['chol'].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
#subplot of chol,age,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True);
#Add data
scatter=ax0.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target'])
#Customise ax0
ax0.set(title="Heart Disease",
       ylabel="Cholestrol");
#Add a legend
ax0.legend(*scatter.legend_elements(),title="Target")
#Add a meanline
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");

#Ax1
scatter=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])
ax1.set(title="Heart Disease",xlabel='Age',ylabel='Thalach')
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(y=over_50['thalach'].mean(),linestyle=':');

#Add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
